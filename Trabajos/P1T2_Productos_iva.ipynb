```python
listproduct = []
listproduct = input("Agregar el precio de los productos separado por comas: ").split(",")
for i in range (len(listproduct)):
  print(f"Valor base del producto ",i+1," : ",listproduct[i])
  print(f"Valor iva (%19) del producto ",i+1," : ",int(listproduct[i])*0.19)
  print(f"Valor total del producto ",i+1," : ",int(listproduct[i])*0.19+int(listproduct[i]),"\n")
```
