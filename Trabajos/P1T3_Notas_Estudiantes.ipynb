```python
from random import randint
cont, cont2 = 0, 0
notas = [randint(10, 50)/10 for i in range(40)]
for j in range (len(notas)):
    if notas[j] <= 2.9:
        cont += 1
    else:
        cont2 += 1
print(f"Perdieron: ", cont,"- Pasaron: ",cont2,"- Promedio fue: ",sum(notas)/40)
print(notas)
```
